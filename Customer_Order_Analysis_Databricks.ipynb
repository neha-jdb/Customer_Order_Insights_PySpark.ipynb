```python
customer_orders = customers.join(
    orders,
    on="customer_id",
    how="inner"
)

customer_orders.show(5)
```

TOTAL ORDERS BY CITY

```python
from pyspark.sql.functions import count

customer_orders.groupBy("city") \
    .agg(count("order_id").alias("total_orders")) \
    .orderBy("total_orders", ascending=False) \
    .show()
```

TOTAL REVENUE BY CITY

```python
from pyspark.sql.functions import sum, round

customer_orders.groupBy("city") \
    .agg(round(sum("total_amount"),2).alias("revenue")) \
    .orderBy("revenue", ascending=False) \
    .show()
```

Top 10 Customers by Spending

```python
from pyspark.sql.functions import sum, round

customer_orders.groupBy(
    "customer_id",
    "name"
).agg(
    round(sum("total_amount"), 2).alias("total_spent")
).orderBy(
    "total_spent",
    ascending=False
).show(10)
```

Order Status Analysis

```python
from pyspark.sql.functions import count

orders.groupBy("status") \
      .agg(count("*").alias("total_orders")) \
      .show()
```

Active vs Inactive Customers

```python
customers.groupBy("is_active") \
         .count() \
         .show()
```

Revenue by Payment Method

```python
order_payment = orders.join(
    payments,
    on="order_id",
    how="inner"
)
```

```python
from pyspark.sql.functions import sum, round

order_payment.groupBy("payment_method") \
             .agg(
                 round(sum("amount"), 2).alias("revenue")
             ) \
             .orderBy("revenue", ascending=False) \
             .show()
```

Revenue by State

```python
from pyspark.sql.functions import sum, round

customer_orders.groupBy("state") \
    .agg(
        round(sum("total_amount"), 2).alias("revenue")
    ) \
    .orderBy("revenue", ascending=False) \
    .show()
```

Average Order Value by City

```python
from pyspark.sql.functions import avg, round

customer_orders.groupBy("city") \
    .agg(
        round(avg("total_amount"), 2).alias("avg_order_value")
    ) \
    .orderBy("avg_order_value", ascending=False) \
    .show()
```

Revenue by Product Category

```python
order_items = orders.join(
    items,
    on="order_id",
    how="inner"
)
```

```python
from pyspark.sql.functions import sum, round

order_items.groupBy("category") \
    .agg(
        round(sum("price"), 2).alias("revenue")
    ) \
    .orderBy("revenue", ascending=False) \
    .show()
```

Most Popular Product Categories

```python
from pyspark.sql.functions import count

order_items.groupBy("category") \
    .agg(
        count("item_id").alias("items_sold")
    ) \
    .orderBy("items_sold", ascending=False) \
    .show()
```

Shipping Method Analysis

```python
order_shipping = orders.join(
    shippings,
    on="order_id",
    how="inner"
)
```

```python
order_shipping.groupBy("shipping_method") \
              .count() \
              .orderBy("count", ascending=False) \
              .show()
```
